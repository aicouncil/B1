```python
q1_age = df2['Age'].quantile(0.25)
q3_age = df2['Age'].quantile(0.75)

print("Quantile 1" , q1_age)
print("Quantile 3" , q3_age)

iqr = q3_age - q1_age
print("Inter Quantile range", iqr)

upper_limit = q3_age + 1.5*iqr
lower_limit = q1_age - 1.5*iqr

print(upper_limit , lower_limit)
```
