**UnSupervised Learning**
   * Clustering -
    - Label the unlabelled data
    - finding out data events with similar properties and form clusters
    - Assign labels to each cluster
    - Statistically analyse each cluster and define significance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({'x':[12,24,28,33,18,29,52,45,24,55,51,61,53,69,72,64,49,58],
                  'y':[36,39,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7]})
```

```python
data
```

```python
plt.scatter(data['x'] , data['y'])
plt.show()
```

* We can assume that we will have 3 clusters but we need to identify the elemnts existing in each cluster

**K-Means algorithm**

```python
#Step 1- Defien 3 random cordinates and call them centeroid
cent1 = [np.random.randint(10,70), np.random.randint(10,70)]
cent2 = [np.random.randint(10,70), np.random.randint(10,70)]
cent3 = [np.random.randint(10,70), np.random.randint(10,70)]

print(cent1, cent2, cent3)
```

```python
plt.scatter(data['x'] , data['y'])

plt.scatter(cent1[0] , cent1[1] , c ='r' , marker='*' , s = 100)
plt.scatter(cent2[0] , cent2[1] , c ='g' , marker='*' , s = 100)
plt.scatter(cent3[0] , cent3[1] , c ='b' , marker='*' , s = 100)

plt.show()
```

```python
print(cent1 , cent2 , cent3)
```

```python
#Step 2 - Calculated distance of each points from all three centeroids one by one

data['r'] = ((data['x'] - cent1[0])**2 + (data['y'] - cent1[1])**2)**0.5        #distances of all points from red color centeroid
data['g'] = ((data['x'] - cent2[0])**2 + (data['y'] - cent2[1])**2)**0.5        #distances of all points from green color centeroid
data['b'] = ((data['x'] - cent3[0])**2 + (data['y'] - cent3[1])**2)**0.5        #distances of all points from blue color centeroid
```

```python
data
```

```python
#Step 3 - find out the centeroid which is closest to each data points
data['labels'] = data[['r','g','b']].idxmin(axis=1)
```

```python
data
```

```python
plt.scatter(data['x'] , data['y'] , c=data['labels'])

#plt.scatter(cent1[0] , cent1[1] , c ='r' , marker='*' , s = 100)
#plt.scatter(cent2[0] , cent2[1] , c ='g' , marker='*' , s = 100)
#plt.scatter(cent3[0] , cent3[1] , c ='b' , marker='*' , s = 100)

plt.show()
```

```python
#Step 4 - Assign new centeroids -Find out average of all the cooredinates in each cluster

cent1 = data[data['labels'] == 'r'][['x','y']].mean().to_list()
cent2 = data[data['labels'] == 'g'][['x','y']].mean().to_list()
cent3 = data[data['labels'] == 'b'][['x','y']].mean().to_list()

print(cent1, cent2, cent3)
```

```python
plt.scatter(data['x'] , data['y'] , c=data['labels'])

plt.scatter(cent1[0] , cent1[1] , c ='r' , marker='*' , s = 100)
plt.scatter(cent2[0] , cent2[1] , c ='g' , marker='*' , s = 100)
plt.scatter(cent3[0] , cent3[1] , c ='b' , marker='*' , s = 100)

plt.show()
```

**Knee rule or elbow method**
   - to decide the number of clusters

```python
data=pd.DataFrame({'x':[12,24,28,33,18,29,52,45,24,55,51,61,53,69,72,64,49,58],
                  'y':[36,39,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7]})
```

```python
plt.scatter(data['x'] , data['y'])
plt.show()
```

```python
inertia = []

for k in range(1,12):
  model_k = KMeans(n_clusters = k)
  model_k.fit(data)
  inertia.append(model_k.inertia_)
```

```python
print(inertia)
```

```python
plt.plot(range(1,12) , inertia , marker='*')
plt.show()
```

```python
model_k_final = KMeans(n_clusters=3)
model_k_final.fit(data)
data['labels'] = model_k_final.predict(data)
```

```python
data
```

```python
plt.scatter(data['x'] , data['y'] , c= data['labels'])
plt.show()
```

```python
model_k_final.cluster_centers_
```

**Case study**
- You own the mall and want to understand the customers like who can be easily converge [Target Customers] so that the sense can be given to marketing team and plan the strategy accordingly.

```python
df = pd.read_csv('https://github.com/bipulshahi/Dataset/raw/refs/heads/main/Mall_Customers.xls')
```

```python
df.head()
```

```python
df1 = df[['Age', 'Annual Income (k$)', 	'Spending Score (1-100)']]
df1.head()
```

```python
'''
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1_scaled = scaler.fit_transform(df1)
'''
```

```python
#Apply elbow method to find optimumn number of clusters

inertia = []
for k in range(1,12):
  model_m = KMeans(n_clusters = k)
  model_m.fit(df1)
  inertia.append(model_m.inertia_)
```

```python
plt.plot(range(1,12) , inertia , marker = '*')
plt.show()
```

```python
#Based on your findings form the clusters and label the data
model_mall = KMeans(n_clusters = 6)
model_mall.fit(df1)
```

```python
#Analyse and define relevance of each clusters
df_cluster_centers = pd.DataFrame(model_mall.cluster_centers_ , columns = df1.columns)
df_cluster_centers
```

```python
df_cluster_centers.plot.bar()
```

```python
#Predictions
df1['labels_predicted'] = model_mall.predict(df1)
```
