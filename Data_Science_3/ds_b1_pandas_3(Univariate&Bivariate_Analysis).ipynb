```python
#'Item_Identifier' vs 'Item_Fat_Content'
pd.crosstab(df['Item_Identifier'] , df['Item_Fat_Content'])
```

```python
pd.crosstab(df['Item_Identifier'] , df['Item_Type'])
```

```python
#continous vs continous
df[['Item_Weight', 'Item_MRP', 'Item_Outlet_Sales']].corr()
```

```python
x1 = df['Item_Weight']
x2 = df['Item_MRP']

n1 = ((x1 - x1.mean()) * (x2 - x2.mean())).sum()

d1 = ((((x1 - x1.mean())**2).sum()) * (((x2 - x2.mean())**2).sum()))**0.5

c1 = n1/d1
print(round(c1,6))
```

```python
x1 = df['Item_MRP']
x2 = df['Item_Outlet_Sales']

n1 = ((x1 - x1.mean()) * (x2 - x2.mean())).sum()

d1 = ((((x1 - x1.mean())**2).sum()) * (((x2 - x2.mean())**2).sum()))**0.5

c1 = n1/d1
print(round(c1,6))
```

```python
x1 = df['Item_Weight']
x2 = df['Item_Outlet_Sales']

n1 = ((x1 - x1.mean()) * (x2 - x2.mean())).sum()

d1 = ((((x1 - x1.mean())**2).sum()) * (((x2 - x2.mean())**2).sum()))**0.5

c1 = n1/d1
print(round(c1,6))
```
