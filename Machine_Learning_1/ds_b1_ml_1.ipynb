**Type 1 - Ordinary Least Square for Simple Linear Regression**

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([3,5,9,7,10])
y = np.array([4,8,9,6,9])

plt.scatter(X,y)
plt.show()
```

```python
#yh = w0 + w1*X

n1 = ((X - X.mean())*(y-y.mean())).sum()
d1 = ((X - X.mean())**2).sum()

w1 = n1/d1
print(w1)

print()

w0 = y.mean() - w1 * X.mean()
print(w0)
```

```python
yh = w0 + w1*X        #calculated output on given X using linear equation components
```

```python
plt.scatter(X,y)

plt.plot(X,yh,'r')

plt.show()
```

```python

plt.xlim(0,12)
plt.ylim(0,12)
plt.scatter(X,y)

plt.plot(X,yh,'r')

plt.show()
```

```python
x_new = 3.8
y_new = w0 + w1 * x_new
print(y_new)
```

```python
error = abs(y - yh).mean()
print(error)
```

**Type 2 - Error optimization process (Gradient Descent)**

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([3,5,9,7,10])
y = np.array([4,8,9,6,9])

plt.scatter(X,y)
plt.show()
```

```python
w0 = 2
w1 = 1

yh = w0 + w1 * X
print(yh)
```

```python
error = ((y - yh)**2).mean()
print(error)
```

```python
dew0 = -2*(y-yh).mean()
print(dew0)

print()

dew1 = -2*((y-yh)*X).mean()
print(dew1)
```

```python
lr = 0.01

w0 = w0 - lr * dew0
w1 = w1 - lr * dew1

print(w0 , w1)
```

```python
yh = w0 + w1 * X

error = ((y - yh)**2).mean()
print(error)
```

```python
w0 = 2
w1 = 1

for i in range(0,20):
  yh = w0 + w1 * X
  error = ((y - yh)**2).mean()
  print(i , error)

  dew0 = -2*(y-yh).mean()
  dew1 = -2*((y-yh)*X).mean()

  lr = 0.01

  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1
```

```python
abs_error = abs(y-yh).mean()
print(abs_error)
```

```python
print(w0 , " " , w1)
```

```python
print(yh)
```

```python
plt.scatter(X,y)
plt.plot(X,yh,'r')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Build apredictive model which can predict sales values**

```python
df = pd.read_csv('https://github.com/bipulshahi/Dataset/raw/refs/heads/main/Advertising.csv',
                 index_col=0)
df.head()
```

```python
X = df['TV']
y = df['sales']

plt.scatter(X,y)
plt.show()
```

```python
w0 = 6
w1 = 6.5

for i in range(0,50):
  yh = w0 + w1 * X

  error = ((y - yh)**2).mean()

  dew0 = -2 * ((y-yh)).mean()
  dew1 = -2 * ((y-yh)*X).mean()

  lr = 0.00001
  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1

  print(i,error)
```

```python
mae = abs(y-yh).mean()
print(mae)
```

```python
plt.scatter(X,y)

plt.plot(X,yh,'r')

plt.show()
```

```python
print(w0 , w1)
```

```python
tv_expense = 54.7
sales_predicted = w0 + w1 * tv_expense
print(sales_predicted)
```

**Multiple Linear Regression**

```python
X = df[['TV' , 'radio']]
y = df['sales']
```

```python
#Yh = w0 + w1*X1 + w2*X2

w0 = 3
w1 = 4
w2 = 5

for i in range(0,1000):
  Yh = w0 + w1*X.TV + w2*X.radio

  error = ((y-Yh)**2).mean()

  dew0 = -2 * ((y - Yh)).mean()
  dew1 = -2 * ((y - Yh)*X.TV).mean()
  dew2 = -2 * ((y - Yh)*X.radio).mean()

  lr = 0.00001
  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1
  w2 = w2 - lr * dew2

  print(i,error)
```

```python
#trained parameters
print(w0 , " " , w1 , "" , w2)
```

```python
#mean absoluter error
mae = abs(y - Yh).mean()
print(mae)
```

```python
#prediction

tv_expense = 54.7
radio_expense = 87.4

sales_predicted = w0 + w1 * tv_expense + w2 * radio_expense
print(sales_predicted)
```
